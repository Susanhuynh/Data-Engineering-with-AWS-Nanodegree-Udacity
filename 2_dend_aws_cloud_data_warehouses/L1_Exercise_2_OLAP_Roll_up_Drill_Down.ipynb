## Roll-up
- Stepping up the level of aggregation to a large grouping
- e.g.`city` is summed as `country`

TODO: Write a query that calculates revenue (`sales_amount`) by day, rating, and country (of a customer). Sort the data by revenue in descending order and limit the data to the top 20 results. The first few rows of your output should match the table below.


```python
%%time
%%sql
SELECT
    dD.day as day,
    dM.rating as rating,
    dC.country as country,
    sum(fS.sales_amount) as revenue
FROM factSales fS
JOIN dimCustomer dC ON fS.customer_key = dC.customer_key
JOIN dimDate dD ON dD.date_key = fS.date_key
JOIN dimMovie dM ON fS.movie_key = dM.movie_key
group by dC.country, dD.day, dM.rating
order by revenue desc
limit 10
```

## Drill-down
- Breaking up one of the dimensions to a lower level.
- e.g.`city` is broken up into  `districts`

TODO: Write a query that calculates revenue (`sales_amount`) by day, rating, and district (of a customer). Sort the data by revenue in descending order and limit the data to the top 20 results. The first few rows of your output should match the table below.


```python
%%time
%%sql
SELECT 
    dimDate.day as day,
    dimMovie.rating as rating,
    dimCustomer.district as district,
    sum(factSales.sales_amount) as revenue
FROM factSales
JOIN dimDate ON factSales.date_key = dimDate.date_key
JOIN dimCustomer ON factSales.customer_key = dimCustomer.customer_key
JOIN dimMovie ON factSales.movie_key = dimMovie.movie_key
group by dimCustomer.district, dimDate.day, dimMovie.rating
order by revenue desc
limit 20
```
